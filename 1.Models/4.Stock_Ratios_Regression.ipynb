```python
def growth (df, col_ratios):

    df = df.sort_values(by=["Stock", "fiscal_year"]).copy()
    df[col_ratios] = df.groupby("Stock")[col_ratios].transform(lambda x: x.fillna(x.mean()))

    def anual_growth (group):
        return group[col_ratios].pct_change(fill_method=None)
    
    df_gr = df.groupby("Stock", group_keys=False).apply(anual_growth )
    df_gr = df_gr.add_suffix("_growth")
    df_gr = df_gr.groupby(df['Stock']).transform(lambda x: x.fillna(x.mean()))
    df = pd.concat([df, df_gr], axis=1)

    return df
```

```python
growth_ratios = ['operating_cash_flow_per_share', 'free_cash_flow_per_share']

df = growth (df, growth_ratios)
df.head()
```

```python
from sklearn.preprocessing import RobustScaler

def preprocess_column(df, columns, winsor_limits=(0.04, 0.04)):
    scaler = RobustScaler()
    
    for column in columns:
    
        df[f"{column}_log"] = np.log1p(df[column] - df[column].min() + 1)
        
        df[f"{column}_wins"] = winsorize(df[f"{column}_log"], limits=winsor_limits)
        
        df[f"{column}_scaled"] = scaler.fit_transform(df[[f"{column}_wins"]])

    return df
```

## Transformation Pipeline

```python
train = train_set.drop(['return'], axis=1)
train_label = train_label["log_return"].copy()
```

```python
excluded_cols = ['Stock', 'adjusted_price', 'close_price', 'period_ending', 'fiscal_period',"days_of_inventory_outstanding", 'inventory_turnover', 'short_term_coverage_ratios',
                    'days_of_payables_outstanding', 'payables_turnover', 'long_term_debt_to_capitalization']

dividend_ratios = ['payout_ratio', 'dividend_payout_ratio', 'dividend_yield']

excluded_cols_2 = excluded_cols + dividend_ratios + ['fiscal_year']

ratios = [col for col in train.columns if col not in excluded_cols_2]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer


def preprocess_column(X):
    X_log = np.log1p(X - X.min() + 1)  
    X_wins = winsorize(X_log, limits=(0.04, 0.04), axis=0)  
    return np.array(X_wins)

ratio_pipeline = make_pipeline(KNNImputer(n_neighbors=5) ,FunctionTransformer(preprocess_column, feature_names_out="one-to-one"), 
                               RobustScaler())

dividend_pipeline = make_pipeline(SimpleImputer(strategy='constant', fill_value=0) ,FunctionTransformer(preprocess_column, feature_names_out="one-to-one"), 
                               RobustScaler())

ord_pipeline = make_pipeline(ordinal_encoder) 

preprocessing = ColumnTransformer([
        ("log_wins_sc", ratio_pipeline, ratios),
        ("log_wins_sc_", dividend_pipeline, dividend_ratios),
        ("ord", ord_pipeline, ['fiscal_year']),
        ("drop_cols", "drop", excluded_cols)
    ],
    remainder='passthrough')
```

```python
train_prepared = preprocessing.fit_transform(train)
train_prepared.shape
```
