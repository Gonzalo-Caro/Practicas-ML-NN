```python
def growth (df, col_ratios):

    df = df.sort_values(by=["Stock", "fiscal_year"]).copy()
    df[col_ratios] = df.groupby("Stock")[col_ratios].transform(lambda x: x.fillna(x.mean()))

    def anual_growth (group):
        return group[col_ratios].pct_change(fill_method=None)
    
    df_gr = df.groupby("Stock", group_keys=False).apply(anual_growth )
    df_gr = df_gr.add_suffix("_growth")
    df_gr = df_gr.groupby(df['Stock']).transform(lambda x: x.fillna(x.mean()))
    df = pd.concat([df, df_gr], axis=1)

    return df
```

```python
growth_ratios = ['operating_cash_flow_per_share', 'free_cash_flow_per_share']

df = growth (df, growth_ratios)
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, stratify=df['return_cat'], test_size=0.2, random_state=42)
```

```python
def income_cat_proportions(data):
    return data["return_cat"].value_counts() / len(data)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(df),
    "Stratified %": income_cat_proportions(test_set),
}).sort_index()

(compare_props * 100).round(2)
```

# Prepara the Data

```python
train = train_set.drop("return", axis=1)
train_label = train_set["return"].copy()
```

```python
outliers_dict = {}

for column in ratios:
    Q1 = train[column].quantile(0.25)
    Q3 = train[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    
    outliers = train[(train[column] < lower_bound) | (train[column] > upper_bound)]
    
    outliers_dict[column] = outliers

    print(f"Outliers in {column}: {outliers.shape[0]} outliers found.")
```

```python
from sklearn.preprocessing import RobustScaler

def preprocess_column(df, columns, winsor_limits=(0.04, 0.04)):
    scaler = RobustScaler()
    
    for column in columns:
    
        df[f"{column}_log"] = np.log1p(df[column] - df[column].min() + 1)
        
        df[f"{column}_wins"] = winsorize(df[f"{column}_log"], limits=winsor_limits)
        
        df[f"{column}_scaled"] = scaler.fit_transform(df[[f"{column}_wins"]])

    return df
```

```python
train = preprocess_column(train, ratios)
```
