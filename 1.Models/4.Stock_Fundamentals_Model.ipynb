# Create a Test Set

```python
import numpy as np

df["return_cat"] = pd.cut(df["return"],
                               bins = [-np.inf, -0.09, 0.10, 0.34, 1.6, np.inf],
                               labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'])
```

```python
plt.figure(figsize=(8, 4))

df["return_cat"].value_counts().sort_index().plot.bar(rot=0, grid=False)
plt.xlabel("Return category")
plt.ylabel("Number of companies")

plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, stratify=df['return_cat'], test_size=0.2, random_state=42)
```

```python
def income_cat_proportions(data):
    return data["return_cat"].value_counts() / len(data)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(df),
    "Stratified %": income_cat_proportions(test_set),
}).sort_index()

(compare_props * 100).round(2)
```

```python
for set_ in (train_set, test_set):
    set_.drop("return_cat", axis=1, inplace=True)
```

```python
train_df = df[df["fiscal_year"] < 2024] 
test_df = df[df["fiscal_year"] == 2024]  

print(f'Train set size:{train_df.shape}, Test set size: {test_df.shape}')
```
