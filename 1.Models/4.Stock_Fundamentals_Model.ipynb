# Create a Test Set

```python
import numpy as np

df["return_cat"] = pd.cut(df["return"],
                               bins = [-np.inf, -0.09, 0.10, 0.34, 1.6, np.inf],
                               labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'])
```

```python
plt.figure(figsize=(8, 4))

df["return_cat"].value_counts().sort_index().plot.bar(rot=0, grid=False)
plt.xlabel("Return category")
plt.ylabel("Number of companies")

plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, stratify=df['return_cat'], test_size=0.2, random_state=42)
```

```python
def income_cat_proportions(data):
    return data["return_cat"].value_counts() / len(data)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(df),
    "Stratified %": income_cat_proportions(test_set),
}).sort_index()

(compare_props * 100).round(2)
```

```python
for set_ in (train_set, test_set):
    set_.drop("return_cat", axis=1, inplace=True)
```

```python
train_df = df[df["fiscal_year"] < 2024] 
test_df = df[df["fiscal_year"] == 2024]  

print(f'Train set size:{train_df.shape}, Test set size: {test_df.shape}')
```

# Discover and Visualize the Data

```python
stocks = train_set.copy()
```

```python
profitability = ["gross_profit_margin", "operating_profit_margin", "return_on_capital_employed", "return"]
efficiency = ["asset_turnover", "inventory_turnover", "receivables_turnover", "operating_cycle", "return"]
liquidity_solvency = ["current_ratio", "quick_ratio", "debt_equity_ratio", "interest_coverage", "return"]
valuation = ["price_to_free_cash_flows_ratio", "price_earnings_ratio", "price_to_book_ratio", "return"]
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(stocks[profitability], figsize=(10, 6))
plt.show()
```

```python
stocks_clean = stocks.copy()
for col in ratios:
    Q1 = stocks_clean[col].quantile(0.25)
    Q3 = stocks_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    stocks_clean = stocks_clean[(stocks_clean[col] >= lower_bound) & (stocks_clean[col] <= upper_bound)]
```

```python
scatter_matrix(stocks_clean[profitability], figsize=(10, 6))
plt.show()
```

```python
scatter_matrix(stocks_clean[efficiency], figsize=(10, 6))
plt.show()
```

```python
scatter_matrix(stocks_clean[liquidity_solvency], figsize=(10, 6))
plt.show()
```

```python
scatter_matrix(stocks_clean[valuation], figsize=(10, 6))
plt.show()
```

```python
import seaborn as sns

ratios_r = ratios + ["return"]

correlation_matrix = stocks[ratios_r].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", fmt=".2f", 
            linewidths=0.5, annot_kws={"size": 10}, vmax=1, vmin=-1)

plt.xticks(rotation=45, fontsize=9)
plt.yticks(rotation=0, fontsize=9)

plt.title('Heatmap: Correlation Matrix')
plt.show()
```

```python
correlation_matrix = stocks_clean[ratios_r].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", fmt=".2f", 
            linewidths=0.5, annot_kws={"size": 10}, vmax=1, vmin=-1)

plt.xticks(rotation=45, fontsize=9)
plt.yticks(rotation=0, fontsize=9)

plt.title('Heatmap: Correlation Matrix')
plt.show()
```
